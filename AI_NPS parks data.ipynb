```python
import json
import requests 
from pprint import pprint
import time
import pandas as pd
from sqlalchemy import create_engine 
from config import api_key
from pandas.io.json import json_normalize
```

```python
sample = 0
url = "https://developer.nps.gov/api/v1/parks?&start="+str(sample)+"&api_key="+api_key
response = requests.get(url).json()
total = int(response["total"])
total
```

```python
park_id = []
park_url = []
fullName = []
parkCode = []
description = []
latitude = []
longitude = []
states = []
directionsInfo = []
directionsUrl = []
weatherInfo = []
designation = []
activities_df = pd.DataFrame()
entranceFees_df = pd.DataFrame()
images_df = pd.DataFrame()

counter = 0
while counter <= total:
    
    url = "https://developer.nps.gov/api/v1/parks?&start="+str(counter)+"&api_key="+api_key
    response = requests.get(url).json()
    parks = response['data']
    for park in parks:
        
        park_id.append(park['id'])    
        park_url.append(park['url'])
        fullName.append(park['fullName'])
        parkCode.append(park['parkCode'])
        description.append(park['description'])
        latitude.append(park['latitude'])
        longitude.append(park['longitude'])
        states.append(park['states'])
        directionsInfo.append(park['directionsInfo'])
        directionsUrl.append(park['directionsUrl'])
        weatherInfo.append(park['weatherInfo'])
        designation.append(park['designation'])

        activities_df = activities_df.append(pd.json_normalize(parks, ["activities"], ["fullName"]))
        entranceFees_df = entranceFees_df.append(pd.json_normalize(parks, ["entranceFees"], ["fullName"]))
        images_df = images_df.append(pd.json_normalize(parks, ["images"], ["fullName"]))
        
    print(url)
    counter = counter +51
    time.sleep(1)
    

    
```

```python
parks_dict = {'park_id':park_id, 'park_url':park_url, 'fullName':fullName,'parkCode':parkCode, 'description':description,\
             'latitude':latitude, 'longitude':longitude, 'states':states, 'directionsInfo':directionsInfo,\
             'directionsUrl':directionsUrl,'weatherInfo':weatherInfo, 'designation':designation}
parks_df = pd.DataFrame(parks_dict)
parks_df.head()
```

```python
activities_df = activities_df.rename(columns={'id':'activity_id', 'name':'activity_name'})
activities_df = activities_df.drop(columns = ['activity_id'])
activities_df
```

```python
entranceFees_df = entranceFees_df.rename(columns={'description':'fee_description', 'title':'fee_type'})
entranceFees_df
```

```python
images_df = images_df.drop(columns = ['title','altText'])
images_df = images_df.rename(columns={'credit':'image_credit', 'caption':'image_caption', 'url':'image_url'})
images_df
```

```python
parks_df.to_json('Resources/nps_park_data.json')
```
