```python
import json
import requests 
from pprint import pprint
import time
import pandas as pd
from sqlalchemy import create_engine 
from config import api_key
from pandas.io.json import json_normalize
```

```python
sample = 0
url = "https://developer.nps.gov/api/v1/webcams?&start="+str(sample)+"&api_key="+api_key
response = requests.get(url).json()
total = int(response["total"])
total
```

```python
response['data']
```

```python
park_code = []
park_webcam_url = []
webcam_title = []
webcam_description = []

counter = 0

while counter <= total:
    
    url = "https://developer.nps.gov/api/v1/webcams?&start="+str(counter)+"&api_key="+api_key
    print('Request Running')
    response = requests.get(url).json()
    parks = response['data']
    for park in parks:
        
        park_webcam_url.append(park['url'])
        webcam_title.append(park['title'])
        webcam_description.append(park['description'])
        cam_park = park['relatedParks']
        a = []
        if cam_park == a:
            park_code.append('Not Available')
        else:
            for code in cam_park:
                code = cam_park[0]['parkCode']
                park_code.append(code)
                
   
    counter = counter +51
    time.sleep(1)
    

    
```

```python
print(len(park_code))
print(len(park_webcam_url))
print(len(webcam_title))
print(len(webcam_description))

```

```python
webcam_dict = {'park_code':park_code, 'park_webcam_url':park_webcam_url,\
               'webcam_title':webcam_title,'webcam_description':webcam_description}
webcam_df = pd.DataFrame(webcam_dict)
webcam_df = webcam_df.reset_index()
webcam_df = webcam_df.rename(columns={'index':'webcam_id'})
webcam_df.head()
```

```python
park_df = pd.read_csv("Resources/nps_park_data.csv")
park_df
```

```python
webcam_df = webcam_df.merge(park_df, how = 'inner', on = 'park_code')
webcam_df = webcam_df[['webcam_id','park_id','park_webcam_url','webcam_title','webcam_description']]
webcam_df
```
