```python
import pandas as pd
import sqlite3
# import sqlalchemy
# from sqlalchemy import create_engine
# from sqlalchemy.ext.automap import automap_base

# engine = create_engine('sqlite:///Resources/Wildfires_Data.sqlite')

# Base = automap_base()
# Base.prepare(engine, reflect=True)
# Base.classes.keys()

# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("Resources/Wildfires_Data.sqlite")
df = pd.read_sql_query("SELECT * from Fires", con)
```

```python
# Verify that result of SQL query is stored in the dataframe
df.head()
```

```python
# Show mulitple specific columns--note the extra brackets
organized_df = df[["OBJECTID", "NWCG_REPORTING_AGENCY", "NWCG_REPORTING_UNIT_NAME", "FIRE_NAME", "FIRE_YEAR", 
                   "STAT_CAUSE_DESCR", "FIRE_SIZE", "FIRE_SIZE_CLASS", "LATITUDE", "LONGITUDE", "STATE"]]
organized_df.head()
```

```python
organized_df.count()
```

```python
# Drop all rows with missing information
clean_df = organized_df.dropna(how='any')
clean_df
```

```python
clean_df.count()
```

```python
# using Loc on the logic test above only returns rows where the result is NPS
nps_df = clean_df.loc[clean_df["NWCG_REPORTING_AGENCY"] == "NPS", :]
nps_df
```

```python
nps_df.count()
```

```python
# The unique method shows every element of the series that appears only once
parks_list = nps_df["NWCG_REPORTING_UNIT_NAME"].unique()
parks_list
```

```python
len(parks_list)
```

```python
# The index can be reset to provide index numbers based on the new rankings.
nps_df = nps_df.reset_index(drop=True)
nps_df.head()
```

```python
# Using .rename(columns={}) in order to rename columns
wildfires_df = nps_df.rename(columns={"NWCG_REPORTING_AGENCY":"park_agency", 
                                      "NWCG_REPORTING_UNIT_NAME":"park_name",
                                      "FIRE_NAME":"fire_name",
                                      "FIRE_YEAR":"fire_year", 
                                      "STAT_CAUSE_DESCR":"cause_description", 
                                      "FIRE_SIZE":"fire_size", 
                                      "FIRE_SIZE_CLASS":"fire_size_class", 
                                      "OBJECTID":"fire_id", 
                                      "LATITUDE": "latitude", 
                                      "LONGITUDE": "longitude", 
                                      "STATE": "state"})
wildfires_df.head()
```
