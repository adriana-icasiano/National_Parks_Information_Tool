```python
import json
import requests 
from pprint import pprint
import time
import pandas as pd
from sqlalchemy import create_engine 
from config import api_key
from pandas.io.json import json_normalize
```

```python
sample = 0
url = "https://developer.nps.gov/api/v1/webcams?&start="+str(sample)+"&api_key="+api_key
response = requests.get(url).json()
total = int(response["total"])
total
```

```python
response['data']
```

```python
park_id = []
park_webcam_url = []
webcam_title = []
webcam_description = []

counter = 0
while counter <= total:
    
    url = "https://developer.nps.gov/api/v1/webcams?&start="+str(counter)+"&api_key="+api_key
    response = requests.get(url).json()
    parks = response['data']
    for park in parks[1:]:
        
        park_id.append(park['id'])    
        park_webcam_url.append(park['url'])
        webcam_title.append(park['title'])
        webcam_description.append(park['description'])
        
    print(url)
    counter = counter +51
    time.sleep(1)
    

    
```

```python
webcam_dict = {'park_id':park_id, 'park_webcam_url':park_webcam_url,\
               'webcam_title':webcam_title,'webcam_description':webcam_description}
webcam_df = pd.DataFrame(webcam_dict)
webcam_df.head()
```

```python
webcam_df.to_csv('Resources/webcam_data.csv', index=False)
```
