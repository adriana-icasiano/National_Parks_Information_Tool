```python
import pandas as pd
import json
```

```python
file = "Resources/species.csv"
```

```python
species = pd.read_csv(file)
species.head()
```

```python
species.drop(species.iloc[:, 9:], inplace=True, axis=1)
```

```python
species.drop(columns = ["Species ID"], inplace=True, axis=1)
```

```python
species.drop(columns = ["Record Status"], inplace=True, axis=1)
```

```python
species.head()
```

```python
species_1 = species["Occurrence"]=="Present"
```

```python
present_species = species[species_1]
```

```python
present_species
```

```python
animals = present_species["Category"].unique().tolist()
animals_1 = animals[0:4]
animals_1
```

```python
animal_species = present_species[present_species["Category"].isin(animals_1)]
animal_species.head()
```

```python
animal_species.drop(columns = ["Occurrence"], inplace=True, axis=1)
```

```python
animal_species
```

```python
animal_species['Park Name'] = animal_species['Park Name'].replace(['Black Canyon of the Gunnison National Park','Denali National Park and Preserve'
                                                                   ,'Gates Of The Arctic National Park and Preserve','Glacier Bay National Park and Preserve',
                                                                  'Great Sand Dunes National Park and Preserve','Haleakala National Park',
                                                                  'Hawaii Volcanoes National Park','Katmai National Park and Preserve','Lake Clark National Park and Preserve','Redwood National Park',
                                                                  'Sequoia and Kings Canyon National Parks'],['Black Canyon Of The Gunnison National Park','Denali National Park & Preserve',
                                                                  'Gates Of The Arctic National Park & Preserve','Glacier Bay National Park & Preserve',
                                                                  'Great Sand Dunes National Park & Preserve','HaleakalƒÅ National Park',
                                                                  "Hawai'i Volcanoes National Park",'Katmai National Park & Preserve',
                                                                  'Lake Clark National Park & Preserve','Redwood National and State Parks','Sequoia & Kings Canyon National Parks'])
```

```python
species_json = (animal_species.groupby(["Park Name"], as_index=True)
                .apply(lambda x: x[["Category", "Order", "Family", "Scientific Name", "Common Names"]].to_dict("r")).reset_index()
                .rename(columns={0: "Animals"})
                .to_json("Resources/park_species.json", orient="records"))
```

```python
animal_species.to_csv("Resources/park_species.csv")
```
