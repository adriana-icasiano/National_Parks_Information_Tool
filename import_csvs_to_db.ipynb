```python
import pandas as pd
from sqlalchemy import create_engine
from config import username, password
import json
import requests 
from pprint import pprint
import time
import pandas as pd
import psycopg2
from sqlalchemy import create_engine 
from flask import jsonify
```

```python
rds_connection_string = f"{username}:{password}@localhost:5432/park_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Produces DB table name as list
engine.table_names()
```

# 1) park table

```python
#Read and display nps_park_data2.csv
park = pd.read_csv("Resources/nps_park_data_final.csv")
#Export data from dataframes to tables
park.to_sql("park", con=engine, index=False, if_exists="replace")
#Read from DB
pd.read_sql_query('select * from park', con=engine).head()
```
